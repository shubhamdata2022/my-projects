# import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data collection

```python
# loading the data set
df= pd.read_csv('happiness_score_dataset.csv')
df
```

```python
# check the n0. of rows and columns in the data set
df.shape
```

```python
# checking for missing values
df.isna().sum()
```

```python
# to simplify the columns name we will rename the columns name
df.rename(columns={'Country':'country_name','Region':'region','Happiness Rank':'happiness_rank','Happiness Score':'happiness_score',
                   'Standard Error':'standard_error','Economy (GDP per Capita)':'gdp_percapita','Family':'family','Health (Life Expectancy)':'health_life_expentency',
                   'Freedom':'freedom','Trust (Government Corruption)':'trust_gov_corr','Generosity':'generosity','Dystopia Residual':'dystopia_residual'},inplace=True)
```

```python
df
```

```python
df.head()
```

# data analyzing

```python
# PLot b/w happiness_score and GDP
plot= plt.figure(figsize=(15,7))
plt.title("Plot between Happiness Score and GDP")
sns.scatterplot (x =df.happiness_score,y=df.gdp_percapita,hue=df.region,s=200)
plt.legend(loc='upper left',fontsize='10')
plt.xlabel('Happiness score')
plt.ylabel ( 'gdp')
```

till here we saw that europ region country has good gdp and good happiness score ,now we will check the all regions country
group their gdp wise

```python
gdp_region=df.groupby('region')['gdp_percapita'].sum()
gdp_region
```

```python
# now plot above data
plot= plt.figure(figsize=(7,7))
gdp_region.plot.pie(autopct = '%1.1f%%')
plt.title( 'GDP by Region')
plt.ylabel( '')
```

```python
# total country
total_country=df.groupby('region')['country_name'].count()
total_country
```

```python
corr = df.corr()
```

```python
#constructing a heatmap to understand the correlation between the columns
plt.figure(figsize=(10,10))
sns.heatmap( corr, cbar=True, square=True, fmt='.1f' ,annot=True, annot_kws= {'size': 8}, cmap= 'Blues')
```

```python
# life expetancy
top_10 = df.head (10)
bottom_10= df.tail(10)
```

```python
fig,axes= plt.subplots(1,2, figsize= (16, 6))
plt.tight_layout(pad= 2)
xlabels= top_10.country_name
axes [0].set_title( 'Top 10 happiest countries Life Expectancy')
axes[0]. set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x= top_10.country_name, y= top_10.health_life_expentency,ax=axes[0])
axes[0].set_xlabel( 'Country Name')
axes[0].set_ylabel('Life expectancy')


xlabels= bottom_10.country_name
axes[1].set_title( 'bottom 10 least happiest countries Life Expectancy')
axes[1].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x= bottom_10.country_name, y= bottom_10.health_life_expentency,ax= axes[1])
axes [1].set_xlabel ('Country Name')
axes [1].set_ylabel('Life expectaney')
```

```python
# plot b/w happiness score and freedom
plot= plt.figure(figsize=(15,7))
plt.title("Plot between Happiness Score and freedom")
sns.scatterplot (x =df.freedom,y=df.happiness_score,hue=df.region,s=200)
plt.legend(loc='upper left',fontsize='12')
plt.xlabel('freedom')
plt.ylabel ( 'Happiness score')
```

```python
# top 10 corrupt country
country = df.sort_values(by= 'trust_gov_corr').head(10)
plot= plt.figure(figsize=(15,7))
plt.title('Countries with Most Perception of Corruption')
plt.xlabel('Country', fontsize= 13)
plt.ylabel ('trust_gov_corr', fontsize = 13)
plt.xticks (rotation = 30, ha='right')
plt.bar(country.country_name, country.trust_gov_corr)
```
