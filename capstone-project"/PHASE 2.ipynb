```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_excel("mydata.xlsx")
df.head()
```

# 2.1 -- Analysis of Literacy rate vs Total crimes.

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='LITRACY_RATE_IN%',y='TOTAL_CRIME',data=df)
```

WE CAN SEE  WHERE LITRACY RATE IS HIGH  THERE CRIME RATE IS LOW...

# 2.2 -- Analysis of Literacy rate vs Each state.

```python
plt.figure(figsize=(12,12))
sns.scatterplot(x='LITRACY_RATE_IN%',y='STATE_UT',data=df)
```

BY OUR ANALYSIS WE CAN CLEARLY SEE THAT KERLA HAS HIGHEST LITRACY RATE AND BIHAR HAS LOWEST LITRACY RATE

# 2.3 -- Analysis of Total Crime vs Each state.

```python
plt.figure(figsize=(12,12))
sns.scatterplot(x='TOTAL_CRIME',y='STATE_UT',data=df)
```

BY OUR ABOVE ANALYSIS WE CAN SEE CRIME RATE IS INVRSELY PROPORTINAL TO  LITRACY RATE FOR MOST OF THE STATES AND UTS CASES , IF LITRACY
RATE IS HIGH THEN CRIME RATE CRIME RATE IS LOW.BUT AREA IS ALSO A USEFUL FACTOR FOR ANALYSING OF CRIME RATE.WE WILL SE IT 
IN BELOW ANALYSIS.

# 2.4 -- Analysis of Area vs Over all Crime.

```python
plt.figure(figsize=(12,12))
sns.scatterplot(x='TOTAL_CRIME',y='AREA_KM_SQURE',data=df)
```

AS WE HAVE ALREADY DISCUSSED AREA IS DIRECTLY PROPORTONAL TO OVERALL CRIME ALSO WE CAN SEE ABOVE PLOT, HERE I PROVED THAT

# 2.5 -- Analysis of Population vs Over all Crime

```python
plt.figure(figsize=(12,12))
sns.scatterplot(x='TOTAL_CRIME',y='POPULATION_IN_THOUSAND',data=df)
```

BY ABOVE ANALYSIS WE CAN SEE IF POPULATION IS HIGH THEN CRIME RATE OF THAT STATE IS ALSO HIGH ,SO POPULATION IS ALSO DIRECTLY
PROPORTINAL TO CRIME RATE , JUST LIKE LAST ANALYSIS
