# 3.1  Insert records from 42_District_ wise_crimes_committed_against_women_2001_2012.csv into a table

in this i have taken record from above excel sheet which i used in sql query and i converted  that exel sheet into csv file  and insert into the database


creating the database and connect to the server in below

```python
import sqlite3
db=sqlite3.connect('crime_against_women.db')
cur=db.cursor()
db.commit()
```

creating the table and column which i will use in sql query

```python
cur.execute('create table womens_crimes(STATE TEXT , DISTRICT TEXT , YEAR INT , RAPE INT , KIDNEPPING INT)')
db.commit()
```

INSERT DATA INTO TABLE

```python
cur.execute("INSERT INTO womens_crimes  VALUES('ANDHRA PRADESH','ADILABAD',2001,50,30)")
```

now checking data stored or not

```python
db.commit()
print(cur.rowcount,"record(s) inserted")
```

now i am going to attach whole data into table in theform of csv format

```python
with open('csv.csv','r') as file:
    no_records = 0
    for row in file:
        cur.execute("INSERT INTO womens_crimes  VALUES (?,?,?,?,?)",row.split(","))
        db.commit()
        no_records +=1
        
print (no_records, 'records inserted')
```

```python
results= cur.execute("select * from womens_crimes")
for row in results:
    print(row)
```

# 3.2 Write SQL query to find the highest number of rapes & Kidnappings that happened in which state, District, and year

i wrote a query which i thought that is ok for above question, may be it would wrong but i gave my best

```python
results=cur.execute('SELECT STATE, DISTRICT, YEAR, MAX(RAPE) AS highest_RAPE,MAX(KIDNEPPING) AS highest_KIDNEPPING FROM womens_crimes GROUP BY STATE, DISTRICT, YEAR ORDER BY highest_RAPE DESC,highest_KIDNEPPING DESC')
for row in results:
    print(row)
```

# 3.3 Write SQL query to find the lowest number of rapes & Kidnappings that happened in which state, District, and year

```python
results=cur.execute('SELECT STATE, DISTRICT, YEAR, MIN(RAPE) AS LOWEST_RAPE,MIN(KIDNEPPING) AS LOWEST_KIDNEPPING FROM womens_crimes GROUP BY STATE, DISTRICT, YEAR ORDER BY LOWEST_RAPE ,LOWEST_KIDNEPPING')
for row in results:
    print(row)
```
